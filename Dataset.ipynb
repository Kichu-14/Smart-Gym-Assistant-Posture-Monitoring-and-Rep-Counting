```python
import cv2
import os
# Define the folder containing the video files
video_folder = "archive/Correct sequence"
# Get the list of video files in the folder
rooms = [os.path.join(video_folder, file) for file in os.listdir(video_folder) if file.endswith(('.mp4', '.avi', '.mkv'))]
# Iterate through each video file in the folder
for i, video_path in enumerate(rooms):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Error: Could not open video source {video_path}")
        continue   
    count = 0
    while True:
        success, image = cap.read()
        if not success:
            print(f"Finished reading video {video_path}")
            break  # Stop reading this video and move to the next one       
        # Check if the image is empty before saving
        if image is None or image.size == 0:
            print(f"Warning: Empty frame from source {video_path} at frame {count}")
            continue
        # Save the frame to file
        cv2.imwrite(f"Dataset/Correct Sequence/room_{i}_frame_{count}.png", image)
        count += 1
    # Release the capture object after done with the video
    cap.release()
cv2.destroyAllWindows()

print("All files have been read.")
```

```python
import cv2
import os

# Define the folder containing the video files
video_folder = "archive/Wrong sequence"

# Get the list of video files in the folder
rooms = [os.path.join(video_folder, file) for file in os.listdir(video_folder) if file.endswith(('.mp4', '.avi', '.mkv'))]

# Iterate through each video file in the folder
for i, video_path in enumerate(rooms):
    cap = cv2.VideoCapture(video_path)
    
    if not cap.isOpened():
        print(f"Error: Could not open video source {video_path}")
        continue
    
    count = 0

    while True:
        success, image = cap.read()

        if not success:
            print(f"Finished reading video {video_path}")
            break  # Stop reading this video and move to the next one
        
        # Check if the image is empty before saving
        if image is None or image.size == 0:
            print(f"Warning: Empty frame from source {video_path} at frame {count}")
            continue

        # Save the frame to file
        cv2.imwrite(f"Dataset/Incorrect Sequence/room_{i}_frame_{count}.png", image)

        count += 1

    # Release the capture object after done with the video
    cap.release()

cv2.destroyAllWindows()

print("All files have been read.")
```
